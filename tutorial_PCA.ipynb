```python
import numpy as np
import numpy.linalg as lin
import matplotlib.pyplot as plt
import pandas as pd
import random
import preprocessing
```

# 데이터 수집

```python
data = np.array([
    [1,2],
    [3,3],
    [3,5],
    [5,4],
    [5,6],
    [6,5],
    [8,7],
    [9,8],
])

# temp = data[:, 1:2].copy()
```

```python
x = data[:,0]
y = data[:,1]

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Oirginal data')
plt.show()
```

# Custom_pca 실행

```python
custom_pca = preprocessing.PCA(data)
```

```python
custom_pca.fit(2)
```

```python
custom_result = custom_pca()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
sklearn_result = pca.fit_transform(data)
```

```python
pca.components_
```

# Custom PCA & Sklearn PCA 방향 계산

```python
from preprocessing import get_angle_in_degrees

s = custom_pca.vector.T
n = 0
angle_list = []
for v1, v2 in zip(s, pca.components_):
    print("Custom vecotr : ", v1)
    print("Sklearn vecotr : ", v2)
    a = get_angle_in_degrees(v1, v2)
    print(f"Custom PCA, sklearn 이루는 각도 : ", a)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], label='Original Data')
plt.scatter(sklearn_result[:, 0], sklearn_result[:, 1], label='Transformed Data')
plt.title('Sklearn PCA on Given Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.savefig('./fig/pca_plot.png')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], label='Original Data')
plt.scatter(custom_result[:, 0], custom_result[:, 1], label='Transformed Data')
plt.title('Custom PCA on Given Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.savefig('./fig/custom_pca_plot.png')
plt.show()
```
